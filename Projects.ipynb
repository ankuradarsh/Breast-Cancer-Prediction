```python
df['diagnosis'] = df['diagnosis'].astype('category')
df['diagnosis'] = df['diagnosis'].cat.codes
```

```python
x = df.drop(columns="diagnosis")
x
```

```python
y = df["diagnosis"]
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import r2_score as rt, accuracy_score, confusion_matrix, classification_report
```

# Split the data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=42)
```

```python
print("Size of the X training data : ", len(x_train))
print("Size of the Y training data : ", len(x_train))
print("Size of the X testing data : ", len(x_test))
print("Size of the Y testing data : ", len(x_test))
print(x.shape, x_train.shape, y_train.shape, x_test.shape, y_test.shape)
```

# Using Decision Tree

```python
model = DecisionTreeClassifier()
```

```python
fitting = model.fit(x_train, y_train)
print("The training score is ", fitting.score(x_train, y_train))
```

```python
predicted = model.predict(x_test)
predicted
```

```python
plt.scatter(y_test, predicted)
plt.xlabel("Actual Trained data")
plt.ylabel("Predicted Trained data")
plt.show()
```

```python
print("The R2 score is : ",rt(y_test, predicted))
print("The accuracy score is : ",accuracy_score(y_test, predicted))
print("The confusion_matrix score is : \n",confusion_matrix(y_test, predicted))
```

```python
report = classification_report(y_test, predicted)
print(report)
```
